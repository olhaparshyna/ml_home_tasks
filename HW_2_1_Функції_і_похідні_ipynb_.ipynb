### –î–æ–º–∞—à–Ω—ñ –∑–∞–≤–¥–∞–Ω–Ω—è –Ω–∞ –≤—ñ–¥–ø—Ä–∞—Ü—é–≤–∞–Ω–Ω—è –ø–æ–Ω—è—Ç—Ç—è —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó

1. –ó–∞–¥–∞–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –æ—Ü—ñ–Ω–∫–∏ –≤—ñ–¥ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω, –≤–∏—Ç—Ä–∞—á–µ–Ω–∏—Ö –Ω–∞ –Ω–∞–≤—á–∞–Ω–Ω—è: $ f(x) = 10x - 5 $.
  1. –ó –¥–æ–ø–æ–º–æ–≥–æ—é Python –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó.
  2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É `matplotlib` –¥–ª—è –ø–æ–±—É–¥–æ–≤–∏ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –∑–Ω–∞—á–µ–Ω—å –≤—ñ–¥ 0 –¥–æ 10 –≥–æ–¥–∏–Ω. –ü—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–π—Ç–µ, —è–∫ –∑–º—ñ–Ω–∞ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω –≤–∏–≤—á–µ–Ω–Ω—è –≤–ø–ª–∏–≤–∞—î –Ω–∞ –æ—Ü—ñ–Ω–∫—É.



   $$
   f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}
   $$
   –ü—ñ–¥—Å—Ç–∞–≤–∏–º–æ $ f(x) = 10x - 5 $:
   $$
   f'(x) = \lim_{h \to 0} \frac{(10(x+h)‚àí5)‚àí(10x‚àí5)}{h}
   $$
   –†–æ–∑–∫—Ä–∏–≤–∞—î–º–æ –¥—É–∂–∫–∏ —Ç–∞ —Å–ø—Ä–æ—â—É—î–º–æ –≤–∏—Ä–∞–∑:
   $$
   f'(x) = \lim_{h \to 0} \frac{10x+10h‚àí5‚àí10x+5}{h}
   $$
   –°–ø—Ä–æ—â—É—î–º–æ:
   $$
   f'(x) = 10
   $$

   –û—Ç–∂–µ –ø–æ—Ö—ñ–¥–Ω–∞ - 10

```python
import numpy as np
import matplotlib.pyplot as plt

# –í–∏–∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó
x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = np.full_like(x, 10)

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10 * x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –ª–µ–≥–µ–Ω–¥–∏, –º—ñ—Ç–æ–∫ —Ç–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞
plt.title("–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

–ú–∞—î–º–æ –Ω–∞ –≥—Ä–∞—Ñ—ñ–∫—É –ª—ñ–Ω—ñ–π–Ω—É —Ñ—É–Ω–∫—Ü—ñ—é —É —è–∫–æ—ó —î —Å—Ç–∞–ª–∞ –ø–æ—Ö—ñ–¥–Ω–∞ 10. –°–ø–æ—Å—Ç–µ—Ä—ñ–≥–∞—î–º–æ, —â–æ –∑—ñ –∑–±—ñ–ª—å—à–µ–Ω–Ω—è–º –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω (–û—Ö) –∑–±—ñ–ª—å—à—É—î—Ç—å—Å—è –∑–Ω–∞—á–µ–Ω–Ω—è –ûy.

2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ Python —ñ `SymPy`, –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—ñ –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ–π $ f(x) = \sin(x) $ —Ç–∞ $ f(x) = \ln(x) $. –ó–æ–±—Ä–∞–∑—ñ—Ç—å —Ü—ñ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—Ö–Ω—ñ –ø–æ—Ö—ñ–¥–Ω—ñ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é `matplotlib`.



## ùëì(ùë•)=sin(ùë•)

```python
x_values = np.linspace(0.1, 10, 400)

f1_values = np.sin(x_values)
f1_prime_values = np.cos(x_values)

plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.plot(x_values, f1_values, label=r'$f(x) = \sin(x)$', color='blue')
plt.plot(x_values, f1_prime_values, label=r"$f'(x) = \cos(x)$", color='green')
plt.title(r"–ì—Ä–∞—Ñ—ñ–∫–∏ –¥–ª—è $f(x) = \sin(x)$ —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show()
```

## ùëì(ùë•)=ln(ùë•)

```python
f2_values = np.log(x_values)
f2_prime_values = 1 / x_values

plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 2)
plt.plot(x_values, f2_values, label=r'$f(x) = \ln(x)$', color='blue')
plt.plot(x_values, f2_prime_values, label=r"$f'(x) = \frac{1}{x}$", color='green')
plt.title(r"–ì—Ä–∞—Ñ—ñ–∫–∏ –¥–ª—è $f(x) = \ln(x)$ —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```


3. –ó–∞–≥–∞–ª—å–Ω—ñ –≤–∏—Ç—Ä–∞—Ç–∏ –∫–æ–º–ø–∞–Ω—ñ—ó –≤—ñ–¥ –æ–±—Å—è–≥—É –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ –∑–∞–¥–∞–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—î—é $ f(x) = x^2 - 5x + 1 $, –¥–µ $ x $ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å. –û–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó –∑ –¥–æ–ø–æ–º–æ–≥–æ—é SymPy —Ç–∞ –≤—ñ–∑—É–∞–ª—ñ–∑—É–π—Ç–µ –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –Ω–∞ –≤—ñ–¥—Ä—ñ–∑–∫—É –≤—ñ–¥ 0 –¥–æ 100.

```python
import sympy as sp

x = sp.symbols('x')
f = x**2 - 5*x + 1

# –û–±—á–∏—Å–ª–µ–Ω–Ω—è –ø–æ—Ö—ñ–¥–Ω–æ—ó
df = sp.diff(f, x)
#2*x - 5
print(df)

# –ì–µ–Ω–µ—Ä–∞—Ü—ñ—è –∑–Ω–∞—á–µ–Ω—å –¥–ª—è –ø–æ–±—É–¥–æ–≤–∏ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤
x_values = np.linspace(0, 100, 100)

# –û–±—á–∏—Å–ª–µ–Ω–Ω—è –∑–Ω–∞—á–µ–Ω—å —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ x
y_values_f = [f.subs(x, val) for val in x_values]
y_values_df = [df.subs(x, val) for val in x_values]

plt.figure(figsize=(10, 6))

plt.plot(x_values, y_values_f, label=r'$f(x) = x^2 - 5x + 1$', color='blue')
plt.plot(x_values, y_values_df, label=r"$f'(x) = 2x - 5$", color='green', linestyle='--')
plt.title("–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó –≤–∏—Ç—Ä–∞—Ç —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel("–ö—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å (x)")
plt.ylabel("–í–∏—Ç—Ä–∞—Ç–∏ (f(x))")
plt.legend()
plt.grid(True)
plt.show()
```
