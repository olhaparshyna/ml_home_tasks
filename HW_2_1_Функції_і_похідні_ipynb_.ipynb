1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



   $$
   f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}
   $$
   Підставимо $ f(x) = 10x - 5 $:
   $$
   f'(x) = \lim_{h \to 0} \frac{(10(x+h)−5)−(10x−5)}{h}
   $$
   Розкриваємо дужки та спрощуємо вираз:
   $$
   f'(x) = \lim_{h \to 0} \frac{10x+10h−5−10x+5}{h}
   $$
   Спрощуємо:
   $$
   f'(x) = 10
   $$

   Отже похідна - 10

```python
import numpy as np
import matplotlib.pyplot as plt

# Визначення функції та її похідної
x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = np.full_like(x, 10)

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10 * x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Маємо на графіку лінійну функцію у якої є стала похідна 10. Спостерігаємо, що зі збільшенням кількості годин (Ох) збільшується значення Оy.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



## 𝑓(𝑥)=sin(𝑥)

```python
x_values = np.linspace(0.1, 10, 400)

f1_values = np.sin(x_values)
f1_prime_values = np.cos(x_values)

plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.plot(x_values, f1_values, label=r'$f(x) = \sin(x)$', color='blue')
plt.plot(x_values, f1_prime_values, label=r"$f'(x) = \cos(x)$", color='green')
plt.title(r"Графіки для $f(x) = \sin(x)$ та її похідної")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show()
```

## 𝑓(𝑥)=ln(𝑥)

```python
f2_values = np.log(x_values)
f2_prime_values = 1 / x_values

plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 2)
plt.plot(x_values, f2_values, label=r'$f(x) = \ln(x)$', color='blue')
plt.plot(x_values, f2_prime_values, label=r"$f'(x) = \frac{1}{x}$", color='green')
plt.title(r"Графіки для $f(x) = \ln(x)$ та її похідної")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
import sympy as sp

x = sp.symbols('x')
f = x**2 - 5*x + 1

# Обчислення похідної
df = sp.diff(f, x)
#2*x - 5
print(df)

# Генерація значень для побудови графіків
x_values = np.linspace(0, 100, 100)

# Обчислення значень функції та її похідної для кожного x
y_values_f = [f.subs(x, val) for val in x_values]
y_values_df = [df.subs(x, val) for val in x_values]

plt.figure(figsize=(10, 6))

plt.plot(x_values, y_values_f, label=r'$f(x) = x^2 - 5x + 1$', color='blue')
plt.plot(x_values, y_values_df, label=r"$f'(x) = 2x - 5$", color='green', linestyle='--')
plt.title("Графік функції витрат і її похідної")
plt.xlabel("Кількість вироблених одиниць (x)")
plt.ylabel("Витрати (f(x))")
plt.legend()
plt.grid(True)
plt.show()
```
